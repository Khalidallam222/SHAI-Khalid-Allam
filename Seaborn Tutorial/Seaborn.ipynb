# Matrix Plot

## Heat maps

```python
# Load the 'car_crashes' dataset from Seaborn
crash_df = sns.load_dataset('car_crashes')
```

```python

# Keep only the numeric columns (excluding the 'insurance' column)
crash_df = crash_df.iloc[:, :-1]

# Set the size of the figure for the heatmap
plt.figure(figsize=(6, 8))

# Set the context and font scale for Seaborn
sns.set_context('paper', font_scale=1.4)

# Calculate the correlation matrix for the numeric variables in the dataset
crash_mx = crash_df.corr()

# Create a heatmap of the correlation matrix
sns.heatmap(crash_mx, annot=True, cmap='Blues', linewidths=1)

# Show the plot
plt.show()
```

Key points about correlation:

1. Range of Values:

- The correlation coefficient, denoted by "r," takes values between -1 and 1.
- A correlation coefficient of 1 indicates a perfect positive linear relationship, where an increase in one variable is associated with a proportional increase in the other.
- A correlation coefficient of -1 indicates a perfect negative linear relationship, where an increase in one variable is associated with a proportional decrease in the other.
- A correlation coefficient of 0 indicates no linear relationship.
2. Direction of Relationship:

- The sign of the correlation coefficient indicates the direction of the relationship.
- A positive correlation (r > 0) indicates that as one variable increases, the other variable tends to increase.
- A negative correlation (r < 0) indicates that as one variable increases, the other variable tends to decrease.
3. Strength of Relationship:

- The magnitude of the correlation coefficient reflects the strength of the relationship.
- The closer the absolute value of the correlation coefficient is to 1, the stronger the linear relationship.
- Values closer to 0 suggest a weaker linear relationship.

```python
# Load the 'flights' dataset from Seaborn
flights = sns.load_dataset('flights')
```

```python

# Display the first few rows of the original dataset
print("Original Dataset:")
print(flights.head())

# Pivot the dataset to create a wide-format DataFrame
flights = flights.pivot_table(index='month', columns='year', values='passengers', observed=False)
# The observed=False means that if there are any missing combinations of years and months, it will still include them in the resulting pivot table.

# Display the first few rows of the pivoted dataset
print("\nPivoted Dataset:")
print(flights.head())

# Create a heatmap to visualize passenger counts over years and months
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)

# Add comments about the parameters in the heatmap function
"""
Parameters in sns.heatmap():
- flights: The DataFrame containing the data to be visualized.
- cmap='magma': The color map to be used in the heatmap. Here, 'magma' is chosen.
- linecolor='white': The color of the lines separating the cells in the heatmap.
- linewidths=1: The width of the lines separating the cells in the heatmap.
"""

# Show the plot
plt.show()
```

# Cluster Map

```python
# Load the 'flights' dataset from Seaborn
flights = sns.load_dataset('flights')
```

```python
# Pivot the dataset to create a wide-format DataFrame
flights_pivot = flights.pivot_table(index='year', columns='month', values='passengers', observed=False)

# Create a clustered heatmap
sns.clustermap(flights_pivot, standard_scale=1, linewidth=1, annot=True)

# Add comments about the parameters in the clustermap function
"""
Parameters in sns.clustermap():
- flights_pivot: The DataFrame containing the data to be visualized.
- cmap='magma': The color map to be used in the heatmap.
- standard_scale=1: Standardize the data along the specified axis (0 or 1) before clustering. 
                   Setting it to 1 standardizes each column (month), resulting in clustering based on patterns across years.
"""

# Show the plot
plt.show()
```

# PairGrid

```python
# Load the 'iris' dataset from Seaborn
iris = sns.load_dataset('iris')
```

```python
# Create a PairGrid for the 'iris' dataset with the 'species' column as the hue
iris_g = sns.PairGrid(iris, hue='species')

# Scatter plots on the upper triangle
iris_g.map_upper(sns.scatterplot)

# Histograms on the diagonal
iris_g.map_diag(sns.histplot)

# KDE plots on the lower triangle
iris_g.map_lower(sns.kdeplot, fill=True)

# Add a legend to differentiate species
iris_g.add_legend()

# Show the PairGrid
plt.show()
```

```python
# Load the 'iris' dataset from Seaborn
iris = sns.load_dataset('iris')
```

```python

# Create a PairGrid for the 'iris' dataset with the 'species' column as the hue
# Specify x_vars and y_vars to define the combinations of variables
iris_g = sns.PairGrid(iris, hue="species",
                      x_vars=["sepal_length", "sepal_width"],
                      y_vars=["petal_length", "petal_width"])

# Use the map method to create scatter plots
iris_g.map(plt.scatter)
```

# Facet Grid

```python
# Load the 'tips' dataset from Seaborn
tips = sns.load_dataset('tips')

# Create a FacetGrid based on the 'time' variable
facet_grid = sns.FacetGrid(tips, col='time', row='sex')

# Map a scatter plot to each subset of the data
facet_grid.map(plt.scatter, 'total_bill', 'tip')

# Add a legend
facet_grid.add_legend()

# Show the plot
plt.show()
```

```python
# Load the 'tips' dataset from Seaborn
tips_df = sns.load_dataset('tips')
```

```python
facet_grid = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3)
facet_grid.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```

```python
# Define size, linewidth and assign a color of white to markers
kws = dict(s=50, linewidth=.5, edgecolor="w")
# Define that we want to assign different markers to smokers and non
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3,
                        hue_order=['Yes','No'], 
                        hue_kws=dict(marker=['^', 'v'], color=['blue', 'orange'], s=[50, 30]))
tips_fg.map(plt.scatter, "total_bill", "tip", **kws)
tips_fg.add_legend()
```

```python
att_df = sns.load_dataset('attention')
att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=4, hue='attention')
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
plt.legend()
plt.show()
```
