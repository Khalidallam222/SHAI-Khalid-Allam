## **1.Quartiles:**
- four Values that divide your data into quarters (4 equal parts), denoted as Q1, Q2, Q3, Q4. Such that:
 - Q1: __25%__ of the data is less than __Q1.__
 - Q2: __50%__ of the data is less than __Q2.__ __It is also known the medien.__
 - Q3: __75%__ of the data is less than __Q3.__
 - Q4: __All__ the data is less than __Q4.__ __Hence, It is also known as the Maximum.__

## **2.Interquartile range (IQR):**
- The difference between the 3rd and 1st quartiles __( IQR = Q3 - Q1 )__.
- Why do we need IQR?
 - Used to build box plots.
 - It is one way to find outliers.

### َQuick example: Find Q1, Q2, Q3, Q4 and the IQR for the following data:

## -15, -9, 2, 3, 3, 4, 4, 6, 7, 7, 7, 8, 9, 11, 12, 24, 30, 40

 - Q1 = 3
 - Q2 = 7
 - Q3 = 11
 - Q4 = 40
 - IQR = 8


# Outliers:
 - https://youtu.be/9jYqZS142mg?t=125
 - __Definition:__ An outlier is an extremely high or low data value compared with the other data values in the dataset.
 - In a sense, this definition leaves it up to the analyst to decide what will be considered abnormal.
 
 - Outliers affect the statistics values of your dataset (like mean and standard deviation)
 - **Statistically**, Outliers can be found using the following formula:
  - **Outlier=** 
    - **Any value smaller than Q1 - 1.5 * IQR.**
    - **Any value bigger than Q3 + 1.5 * IQR.**

# Boxplot:
 - A box plot shows the five number summary of a dataset:
   - __Minimum value.__
   - __Lower quartile (Q1).__
   - __Median (Q2).__
   - __Upper quartile (Q3).__
   - __Maximum value.__


### You can draw a box plot for a feature in your dataset using the seaborn function 'boxplot'.
#### sns.boxplot(x= )
https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
sns.boxplot(data=iris, orient="h", palette="Set2")
plt.show()
```
