# **Preprocessing**

```python
df.isnull().sum()
```

```python
train_set.columns
```

```python
X_train = train_set.drop(columns=['price'])
y_train = train_set['price']
X_train2 = X_train
```

```python
df.head()
```

```python
cat_data =X_train[cat_col]
cat_data
```

```python
ohe = OneHotEncoder()
cat_data_encoded = ohe.fit_transform(cat_data)
cat_data_encoded=cat_data_encoded.toarray()
cat_data_encoded.shape
```

```python
num_data = X_train.select_dtypes(include= np.number)
num_data
```

```python
ss = StandardScaler()
num_data_scaled= ss.fit_transform(num_data)
num_data_scaled
```

```python
combined_array = np.hstack((num_data_scaled,cat_data_encoded))
combined_df = pd.DataFrame(combined_array)
```

```python
column_names = list(ohe.get_feature_names_out())+num_data.columns.to_list()
combined_df.columns = column_names
```

```python
combined_df
```

```python
cat_pipeline = Pipeline(
    [("encoder",OneHotEncoder())]
)
num_pipeline = Pipeline(
    [("scaler", StandardScaler())]
)
```

```python
preprocessing = make_column_transformer(
    (num_pipeline,
    make_column_selector(dtype_include=np.number)),
    (cat_pipeline,
    make_column_selector(dtype_include=object))
)
```

```python
train_set_prepared = preprocessing.fit_transform(X_train2)
```

# **Modeling**

```python
models=[]
models_rmse=[]
```



---
## simple regression model

---



```python
lr = LinearRegression()
lr.fit(train_set_prepared,y_train)
```

```python
lin_rmse = mean_squared_error(y_train , lr.predict(train_set_prepared), squared=False)
models.append('lr')
models_rmse.append(lin_rmse)
lin_rmse
```



---
##  one decision tree

---



```python
dtr = DecisionTreeRegressor(random_state=42)
dtr.fit(train_set_prepared,y_train)
dtr_rmse = -cross_val_score(dtr, train_set_prepared,y_train,cv = 10 , scoring= 'neg_root_mean_squared_error')
```

```python
dtr_rmse.mean()
```

```python
models.append('dtr')
models_rmse.append(dtr_rmse.mean())
```



---
## random forest

---



```python
rfr = RandomForestRegressor()
rfr.fit(train_set_prepared,y_train)
rfr_rmse = -cross_val_score(rfr, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
rfr_rmse.mean()
```

```python
models.append('random forest')
models_rmse.append(rfr_rmse.mean())
```

```python
degree = 3
poly  = PolynomialFeatures(degree= degree)
X_train_poly=poly.fit_transform(train_set_prepared)

poly_model = LinearRegression()
poly_model.fit(X_train_poly,y_train)

poly_rmse = -cross_val_score(poly_model, train_set_prepared,y_train,cv=3,scoring= 'neg_root_mean_squared_error')
poly_rmse.mean()
```



---
## xgboost regressor

---



```python
xgb_model = xgb.XGBRegressor(objective="reg:linear")
xgb_model.fit(train_set_prepared,y_train)
xgb_rmse = -cross_val_score(xgb_model, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
```

```python
xgb_rmse.mean()
```

```python
models.append('xgboost')
models_rmse.append(xgb_rmse.mean())
```


---
## gradiant boosting

---



```python
gb_model = GradientBoostingRegressor(random_state=0).fit(train_set_prepared, y_train)
gb_rmse = -cross_val_score(gb_model, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
```

```python
gb_rmse.mean()
```

```python
models.append('gradiant boosting')
models_rmse.append(gb_rmse.mean())
```

---
## KNeighborsRegressor

---

```python
kn_model = KNeighborsRegressor().fit(train_set_prepared, y_train)
kn_rmse = -cross_val_score(kn_model, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
```

```python
kn_rmse.mean()
```

```python
models.append('Knearest regressor ')
models_rmse.append(kn_rmse.mean())
```

---
## Ridge Regression

---

```python
ridge_reg = Ridge().fit(train_set_prepared, y_train)
ridge_rmse = -cross_val_score(ridge_reg, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
ridge_rmse.mean()
```

```python
models.append('ridge regressor ')
models_rmse.append(ridge_rmse.mean())
```

---
## lasso Regression

---

```python
lasso_reg = Lasso().fit(train_set_prepared, y_train)
lasso_rmse = -cross_val_score(lasso_reg, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
lasso_rmse.mean()
```

```python
models.append('lasso regressor ')
models_rmse.append(lasso_rmse.mean())
```

---
## Elastic Net Regression

---


```python
net_reg = ElasticNet().fit(train_set_prepared, y_train)
net_rmse = -cross_val_score(net_reg, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
net_rmse.mean()
```

```python
models.append('Elastic Net regressor ')
models_rmse.append(net_rmse.mean())
```

---
## Adaboost

---


```python
ada_reg = AdaBoostRegressor(n_estimators = 100).fit(train_set_prepared, y_train)
ada_rmse = -cross_val_score(ada_reg, train_set_prepared,y_train,cv=10,scoring= 'neg_root_mean_squared_error')
ada_rmse.mean()
```

```python
models.append('ada_reg regressor ')
models_rmse.append(ada_rmse.mean())
```

```python
models
```

# let's look at the models' rmse

```python
models_scores=pd.DataFrame({"model":models, "rmse_score":models_rmse})
models_scores.sort_values(by="rmse_score")
```
