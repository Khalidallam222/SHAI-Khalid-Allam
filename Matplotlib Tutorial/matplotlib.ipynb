# Basic Graph

```python
# Sample data
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize the figure
plt.figure(figsize=(8, 6), dpi=300)

## Line 1
# plt.plot(x,y, label='2x', color='green', linewidth=3, marker='.', markersize=10, markeredgecolor='red', alpha=0.5, linestyle='--')

## Use shorthand notation
## fmt = '[color][marker][line]'
plt.plot(x,y, 'go-', label='2x')

# Line 2: Plotting x^2 in red, with solid line up to x=3, then dashed line
x2 = np.arange(0, 4, 0.5)
plt.plot(x2[:4], x2[:4] ** 2, 'r', label='x^2')
plt.plot(x2[3:], x2[3:] ** 2, 'r--')


# Adding title and axis labels
plt.title("Our First Graph", fontdict={'fontname': 'TIMES NEW ROMAN', 'fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')


# Setting custom ticks
plt.xticks([1, 2, 3, 4])
plt.yticks([2, 4, 5, 7])

# Displaying legend
plt.legend()

# Saving the figure
plt.savefig('myFirstGraph.png', dpi=300)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example 1: Solid red line with circles as markers
plt.plot(x, y, 'ro-')

# Example 2: Dashed blue line with squares as markers
plt.plot(x, y, 'bs--')

# Example 3: Green dashed line without markers
plt.plot(x, y, 'g--')

# Example 4: Black dotted line with triangles as markers
plt.plot(x, y, 'k^:')

plt.show()
```

# Bar Chart

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 35, 30, 25]

# Create a bar chart with hatch patterns
colors = ['blue', 'green', 'orange', 'red']
hatch_patterns = ['/', '\\', '|', '-']
plt.bar(categories, values, color=colors, alpha=0.7, label=['A','B','C','D'], hatch=hatch_patterns)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
## Use set_hatch

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 35, 30, 25]

# Create a bar chart
bars = plt.bar(categories, values, color='blue', alpha=0.7)

# Specify hatch patterns for each bar
hatch_patterns = ['/', '\\', '|', '-']

# Set hatch patterns for each bar
for bar, hatch_pattern in zip(bars, hatch_patterns):
    bar.set_hatch(hatch_pattern)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Custom Hatch Patterns')

# Display the plot
plt.show()
```

### Example1: Gas data

```python
# Read the CSV file into a DataFrame
gas = pd.read_csv('gas_prices.csv')

# Set the figure size for the plot
plt.figure(figsize=(15, 5))

# Set the title of the plot
plt.title('Gas Prices over Time (in USD)')

# Plot gas prices for different countries over the years
plt.plot(gas['Year'], gas['Australia'], '.-', label='Australia')
plt.plot(gas['Year'], gas['Canada'], '.-', label='Canada')
plt.plot(gas['Year'], gas['South Korea'], '.-', label='South Korea')

# Set the x-axis ticks to be the years
plt.xticks(gas['Year'])

# Set labels for y-axis and x-axis
plt.ylabel('USD')
plt.xlabel('Year')

# Display a legend to identify each country's data
plt.legend()

# Show the plot
plt.show()
```

```python
# Read the CSV file into a DataFrame
gas = pd.read_csv('gas_prices.csv')

# Set the figure size for the plot
plt.figure(figsize=(15, 5))

# Set the title of the plot
plt.title('Gas Prices over Time (in USD)')

# Get column names
cols = gas.columns.values

# Plot gas prices for all countries over the years
# Using iloc[:, 1:] to select all columns starting from the second one
plt.plot(gas['Year'], gas.iloc[:, 1:], '.-', label=cols[1:])

# Set the x-axis ticks to be the years
plt.xticks(gas['Year'])

# Set labels for y-axis and x-axis
plt.ylabel('USD')
plt.xlabel('Year')

# Display a legend with labels from the column names
plt.legend(bbox_to_anchor=(1, 1))

# Save the figure as an image file
plt.savefig('gas_prices_figure', dpi=300)

# Show the plot
plt.show()
```

# Pie Chart

```python
# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]
colors = ['blue', 'orange', 'green', 'red']

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%.1f%%', startangle=90)

# Set aspect ratio to be equal, making the pie chart a circle
plt.axis('equal')

# Add a title
plt.title('Distribution of Categories')

# Display the pie chart
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]

plt.pie([left,right], labels=['Left', 'Right'], colors=['yellow','green'], autopct='%.1f%%')
plt.title('Preferred Foot Percentage')
plt.show()
```

```python
# Extract numeric part of the 'Weight(lbs)' column (execute this one time)
# fifa['Weight'] = fifa['Weight'].str.extract('(\d+)').astype(float)

# Convert 'Weight(lbs)' to 'Weight(kg)'fifa (execute this one time)
# fifa['Weight'] = fifa['Weight'] / 2.20462

# Set the ggplot style
plt.style.use('ggplot')

# data 
data = fifa['Weight']

# Create a pie chart
plt.figure(figsize=(5, 5))

# Compute the histogram counts and multiply by the total number of data points
counts = plt.hist(data, bins=5)[0] * len(data)

# Labels for each category
labels = ['50-60 kg', '60-75 kg', '75-85 kg', '85-100 kg', '100-110 kg']

# Define explode values for each category
explode = (0.5, 0, 0, 0.2, 0.4)

# Plot the pie chart
plt.pie(counts, labels=labels, autopct='%.1f%%', pctdistance=0.8, explode=explode)

# Add a title to the pie chart
plt.title('Weight Distribution')

# Show only the pie chart
plt.show()
```

```python
plt.style.use('default')
Barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
Madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']

plt.boxplot([Barcelona, Madrid], labels=['FC Barcelona', 'Real Madrid'])
plt.show()
```

```python
Teams = fifa.groupby(['Club'])
sub_Teams = list(Teams.groups.keys())[0:5]

plt.figure(figsize=(15,5))
colors = ['red', 'green', 'orange', 'yellow']
for i, team in enumerate(sub_Teams):
    t = Teams.get_group((team,))['Overall']
    boxprops = dict(color=colors[i%4], linewidth=2)
    plt.boxplot(t, labels=[team], positions=[i], boxprops=boxprops)
    

plt.show()
    
```
